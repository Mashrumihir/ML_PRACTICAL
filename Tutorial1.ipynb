# Exercise 1: Describe Any 10 AI Libraries

## Description
This tutorial provides an overview of **10 popular Artificial Intelligence libraries**, focusing on their **functionalities, applications, and key features**.

---

## 1. TensorFlow
**TensorFlow** is an open-source machine learning library developed by Google. It is mainly used for building deep learning models.

### Features
- Supports neural networks and deep learning  
- Works on CPU, GPU, and TPU  
- Scalable and flexible  

### Applications
- Image recognition  
- Speech processing  
- Natural language processing  

---

## 2. PyTorch
**PyTorch** is an open-source deep learning framework developed by Facebook AI Research.

### Features
- Dynamic computation graphs  
- Easy debugging  
- Python-based and flexible  

### Applications
- Deep learning research  
- Computer vision  
- NLP tasks  

---

## 3. Keras
**Keras** is a high-level neural network library that works on top of TensorFlow.

### Features
- Simple and easy to use  
- Faster model development  
- Good for beginners  

### Applications
- Image classification  
- Neural network modeling  
- Text processing  

---

## 4. OpenCV
**OpenCV** is an open-source library mainly used for computer vision and image processing.

### Features
- Real-time image processing  
- Supports multiple programming languages  
- Large collection of vision algorithms  

### Applications
- Face detection  
- Object tracking  
- Video analysis  

---

## 5. Scikit-Learn
**Scikit-Learn** is a Python library for machine learning and data analysis.

### Features
- Simple APIs  
- Supports supervised and unsupervised learning  
- Built on NumPy and SciPy  

### Applications
- Classification  
- Regression  
- Clustering  

---

## 6. NLTK (Natural Language Toolkit)
**NLTK** is a library used for processing and analyzing human language data.

### Features
- Text preprocessing tools  
- Supports tokenization and stemming  
- Includes linguistic datasets  

### Applications
- Sentiment analysis  
- Text classification  
- Language processing  

---

## 7. SpaCy
**SpaCy** is a modern NLP library designed for real-world applications.

### Features
- Fast and efficient  
- Pre-trained language models  
- Industrial-grade NLP  

### Applications
- Named Entity Recognition (NER)  
- Dependency parsing  
- Chatbot development  

---

## 8. GPT-3 (OpenAI)
**GPT-3** is a powerful language model developed by OpenAI.

### Features
- Generates human-like text  
- Uses deep learning transformers  
- Few-shot learning capability  

### Applications
- Chatbots  
- Content generation  
- Code assistance  

---

## 9. IBM Watson
**IBM Watson** is an AI-based cloud platform that provides intelligent services.

### Features
- Speech and text processing  
- Machine learning services  
- Cloud-based AI solutions  

### Applications
- Healthcare analytics  
- Business intelligence  
- Customer service bots  

---

## 10. RapidMiner
**RapidMiner** is a data science and machine learning platform.

### Features
- Visual workflow design  
- No or low coding required  
- Strong data preprocessing tools  

### Applications
- Predictive analytics  
- Data mining  
- Business decision support  

---

## Conclusion
These AI libraries help developers and researchers build intelligent systems efficiently. They are widely used in academics and industries for solving real-world problems.

1. Count plots (used for categorial data)

```python
# 10 Draw Count plot for â€˜classâ€™.
sns.countplot(y='source_city',data=df)
```

2. Piechart (used for categorial data)
- A pie chart represents the parts of a whole entity.

```python
# 11 Draw a pie plot for â€˜source_cityâ€™.
df['source_city'].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

Donut plots (used for categorical data)
Dount plots are used to show the proportion of each category in relation to the whole.

```python
# 12 Draw Donut plot for â€˜source_cityâ€™.
plt.pie(df['source_city'].value_counts(), labels=df['source_city'].unique(), autopct='%1.2f%%')
```

```python
plt.pie(df['source_city'].value_counts(), labels=df['source_city'].unique(), autopct='%1.2f',wedgeprops=dict(width=0.7))
```

3. Histogram (used for numerical data) (continous variables)
- Histogram are used when you a continuous variables, and you want to visualiaze its distribution.

```python
# 13 Draw Histogram plot for â€˜priceâ€™.
plt.hist(df['price'])
```
